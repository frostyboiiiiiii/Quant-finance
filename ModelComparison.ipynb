```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
def black_scholes(S, K, T, r, sigma, option_type="call"):

    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)

    if option_type == "call":
        price = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    else:
        price = K * np.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)

    return price
```

```python
def monte_carlo(S, K, T, r, sigma, option_type = 'call', simulations = 100000):

    Z = np.random.standard_normal(simulations)
    S_T = S * np.exp((r - 0.5 * sigma**2) * T + sigma * np.sqrt(T) * Z)

    if option_type == 'call':
        payoffs = np.maximum(S_T - K, 0)
    else:
        payoffs = np.maximum(K - S_T,0)

    price = np.exp(-r * T) * np.mean(payoffs)
    return price
```

```python
S, K, T, r, sigma = 100, 100, 1, 0.05, 0.2

bs_price = black_scholes(S, K, T, r, sigma, option_type='call')
mc_price = monte_carlo(S, K, T, r, sigma, option_type='call', simulations=100000)

print(f"Black-Scholes: {bs_price:.4f}")
print(f"Monte Carlo:   {mc_price:.4f}")
print(f"Difference:    {abs(bs_price - mc_price):.4f}")
```
